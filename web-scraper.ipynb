```python
def get_scraped_values(result,tag, class_name, attribute):
    try:
        val = result.find(tag, {class_name:attribute}).text
        return val
    except AttributeError:
        return None
```

```python
def return_scraped_values(result):
    address = get_scraped_values(result,"p","data-testid","address")
    floor_area = get_scraped_values(result,"p","data-testid","floor-area")
    agent_name = get_scraped_values(result,"p","data-testid","agent-name")
    price = get_scraped_values(result,"div","data-testid","price")
    baths = get_scraped_values(result, "p", "data-testid", "baths")
    beds = get_scraped_values(result, "p", "data-testid", "beds")
    data = [address, floor_area, agent_name, baths, beds, price]
    return data
```
